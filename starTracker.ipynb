```python
from IPython.core.display import Image 



#ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)  
#plt.imshow(thresh,'gray')
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('star.jpg',0)
img_spare = img
# Initiate FAST object with default values
fast = cv2.FastFeatureDetector()

# find and draw the keypoints
kp = fast.detect(img,None)
img2 = cv2.drawKeypoints(img, kp, img_spare, color=(255,0,0))
plt.imshow(img2)
# Print all default params
fast.setInt('threshold',100)
print "Threshold: ", fast.getInt('threshold')
print "nonmaxSuppression: ", fast.getBool('nonmaxSuppression')
#print "neighborhood: ", fast.getInt('type')
print "Total Keypoints with nonmaxSuppression: ", len(kp)
```

```python
cv2.imwrite('fast_true.png',img2)
thresh = 70
# Disable nonmaxSuppression
fast.setBool('nonmaxSuppression',0)
while len(kp) > 500:

    fast.setInt('threshold',thresh)
    #print "Threshold: ", fast.getInt('threshold')

    kp = fast.detect(img,None)
    #print "Total Keypoints without nonmaxSuppression: ", len(kp)
    thresh += 10
while len(kp) < 500:

    fast.setInt('threshold',thresh)
    #print "Threshold: ", fast.getInt('threshold')

    kp = fast.detect(img,None)
    #print "Total Keypoints without nonmaxSuppression: ", len(kp)
    thresh -= 1
print len(kp)
img3 = cv2.drawKeypoints(img, kp, img_spare, color=(255,0,0))
plt.figure(figsize=(12,12))
plt.imshow(img3)
cv2.imwrite('fast_false.png',img3)
```

```python
newkp = []
flag = False
for keypoint1 in kp:
    for keypoint2 in newkp:
        dist = ((keypoint1.pt[0]-keypoint2.pt[0])**2+(keypoint1.pt[1]-keypoint2.pt[1])**2)**0.5
        if dist < 10 and dist > 0:
            flag = True
            break
    if flag == False:
        newkp.append(keypoint1)
        #print "appended"
    flag = False
#for keypoint in newkp:    
    #print "x:" +str(keypoint.pt[0])+" y:"+str(keypoint.pt[1])
    
img4 = cv2.drawKeypoints(img, newkp, img_spare, color=(255,0,0))
plt.figure(figsize=(12,12))
plt.imshow(img4)
print "Stars Detected:" + str(len(newkp))
```
